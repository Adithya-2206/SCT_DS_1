```python
from google.colab import files

uploaded = files.upload()
import pandas as pd

df = pd.read_csv('matches.csv')

print(df.head())
print(df.columns)
import matplotlib.pyplot as plt

# Corrected line to use 'result_margin' and filter for 'runs'
run_margins = df[(df['result'] == 'runs') & (df['result_margin'] > 0)]['result_margin']

plt.figure(figsize=(10, 6))
plt.hist(run_margins, bins=20, edgecolor='black')

plt.title('Distribution of IPL Winning Margins (Runs)')
plt.xlabel('Winning Margin (Runs)')
plt.ylabel('Frequency')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
print(df.columns.tolist())
```
